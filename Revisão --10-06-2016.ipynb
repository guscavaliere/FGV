```python
import requests
from bs4 import BeautifulSoup
```

```python
url ="http://christiananswers.net/bible/gen1.html"
page=requests.get(url)
html=page.content
sopa=BeautifulSoup(html,"lxml")
```

```python
texto = sopa.text
```

```python
lista_palavras=texto.split()
```

```python
lista_palavras
```

```python
import string
```

```python
for s in string.punctuation:
    texto =texto.replace(s," ")
```

```python
for s in string.digits:
    texto =texto.replace(s," ")
```

```python
lista_palavras=texto.split()
```

```python
lista_palavras
```

```python
maiusculas=[p for p in lista_palavras if p.istitle()]
```

```python
maiusculas
```

```python
from collections import Counter
```

```python
freqdist=Counter(maiusculas)
freqdist.most_common(100)
```
